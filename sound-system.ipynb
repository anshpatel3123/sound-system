```python
import math
import numpy as np
import matplotlib.pyplot as plt
from thinkdsp import Wave, read_wave
pi = math.pi
import sounddevice as sd
```

```python
fs = 11025  # Sample rate
seconds = 10  # Duration of recording

recording = sd.rec(int(seconds * fs), samplerate=fs, channels=1)
sd.wait()  # Wait until recording is finished
```

```python
xs = recording[:,0]
xwave = Wave(xs,framerate=11025)
xwave.plot()
```

```python
xwave.make_audio();
```

```python
tzero = 5000
ys = xs[tzero:tzero+2**16]
ywave = Wave(ys, framerate=11025)

ywave.plot()
ywave.make_audio()
```
